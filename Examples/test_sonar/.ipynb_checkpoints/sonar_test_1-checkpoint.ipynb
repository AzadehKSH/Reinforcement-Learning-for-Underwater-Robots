```python
import holoocean
import numpy as np
import json
import math
import sys
import matplotlib.pyplot as plt


from IPython.display import display
from IPython.display import clear_output
import IPython
```

```python
with open("./config.json", 'r') as f:
    cfg = json.load(f)
```

```python
config = cfg['agents'][0]['sensors'][-1]["configuration"]
azi = 120
minR = 0.5
maxR = config['RangeMax']
binsR = 750
binsA = 480
```

```python
%matplotlib inline
plt.ion()
fig, ax = plt.subplots(subplot_kw=dict(projection='polar'), figsize=(8,5))
ax.set_theta_zero_location("N")
ax.set_thetamin(-azi/2)
ax.set_thetamax(azi/2)

theta = np.linspace(-azi/2, azi/2, binsA)*np.pi/180
r = np.linspace(minR, maxR, binsR)
T, R = np.meshgrid(theta, r)
z = np.zeros_like(T)

plt.grid(False)
plot = ax.pcolormesh(T, R, z, cmap='gray', shading='auto', vmin=0, vmax=1)
plt.tight_layout()
fig.set_facecolor("w")
fig.canvas.flush_events()
```

```python
env = holoocean.make(scenario_cfg=cfg)
```

```python
action = np.zeros(8)
action[4:8] += 200
```

```python
%matplotlib inline
states = env.tick()
env_map = []
for i in range(20):
    env.act('auv0', action)
    states = env.tick(num_ticks=20)
    vel = states["DVLSensor"]
    print("Hi")
    clear_output()
    if 'ProfilingSonar' in states:
            s = states['ProfilingSonar']
            plot.set_array(s.ravel())
            fig.canvas.draw()
            fig.canvas.flush_events()
            display(fig)
    sonar = states["ProfilingSonar"]
    display(len(sonar))
    display(states["ProfilingSonar"])
#    display(action)
    
```

```python
def f(r, theta):
    return r * np.sin(theta), r * np.cos(theta)
```

```python
plot.set_array(a.ravel())
fig.canvas.draw()
fig.canvas.flush_events()
display(fig)
```

```python
tmp_coords = []

for index_1, i in enumerate(a):
    if i.any():
        # array has at least one value other than zero
        for index_2, j in enumerate(i):
            if j == 0.: continue
            coords = f(R[index_1][index_2], T[index_1][index_2])
            if coords != (0.,0.):
                tmp_coords.append(coords)
#display(tmp_coords)
```
